```python
def twoSumBruteForce(nums, target):
    # https://leetcode.com/problems/two-sum/description/
    for i in range(len(nums)):
        for j in range(i+1, len(nums)):
            if nums[i] + nums[j] == target:
                return i, j
    return None

def twoSum(nums, target):
    d = {}
    for k, v in enumerate(nums, 0):
        if (target - v) in d:
            return (d[target-v], k)
        else:
            d[v] = k

print(twoSumBruteForce([2, 5, 3, 6], 8))
print(twoSum([2, 5, 3, 6], 8))   
```

```python
def containsDuplicateBruteForce(nums):
    # https://leetcode.com/problems/contains-duplicate/
    for i in range(len(nums)):
        for j in range(i+1, len(nums)):
            if nums[i] == nums[j]:
                return True
    return False        

def containsDuplicate(nums):
    d = {}
    for i in nums:
        if i not in d:
            d[i] = 1
        else:
            return True
    return False   
print(containsDuplicate([1, 2, 3, 4, 5, 1]))        
print(containsDuplicateBruteForce([1, 2, 3, 4, 5]))        
```

```python
def validAnagram(s, t):
    # https://leetcode.com/problems/valid-anagram/
    s_map, t_map = {}, {}
    for i in s:
        if i not in s_map:
            s_map[i] = 1
        else:
            s_map[i] += 1
    for j in t:
        if j not in t_map:
            t_map[j] = 1
        else:
            t_map[j] +=1     

    return s_map == t_map  

print(validAnagram('bala','alab'))          
```

```python
def twoPointerTwoSum(nums, target):
    # https://leetcode.com/problems/two-sum-ii-input-array-is-sorted/
    left, right = 0, len(nums)-1

    while left < right:
        if nums[left] + nums[right] == target:
            return left, right
        elif nums[left] + nums[right] > target:
            right = right - 1
        else: 
            left = left+1
print(twoPointerTwoSum([2,5,7,9],12))            
```

```python
def romanToInt(s):
    # https://leetcode.com/problems/roman-to-integer/
    d = {'I':1,'V':5,'X':10,'L':50,'C':100,'D':500,'M':1000}
    sum = 0

    for i in range(len(s)):
        if i+1 < len(s) and d[s[i]] < d[s[i+1]]:
            sum = sum - d[s[i]]
        else:
            sum = sum + d[s[i]]

    return sum
print(romanToInt('MCMXCIV'))    

```

```python
def majorityElement(nums):
    # https://leetcode.com/problems/majority-element/
    n = len(nums)//2
    d = {}
    for i in nums:
        if i not in d:
            d[i] = 1
        else:
            d[i] += 1
    for k, v in d.items():
        if v >= n:
            return k     

print(majorityElement([1, 3, 3, 3, 3, 6, 7]))
```

```python
def lengthOfLastWord(s):
    # https://leetcode.com/problems/length-of-last-word/
    sStrip = s.split()
    return (len(sStrip[-1]))

def lengthOfLastWord(inputString: str) -> int:
        
    # Initialize the length of the last word to zero
    lastWordLength = 0
        
    # Start from the end of the string
    index = len(inputString) - 1
        
    # Skip trailing spaces
    while index >= 0 and inputString[index] == ' ':
        index -= 1
        
    # Count the length of the last word
    while index >= 0 and inputString[index] != ' ':
        lastWordLength += 1
        index -= 1
        
    # Return the length of the last word
    return lastWordLength    

print(lengthOfLastWord("   fly me   to   the moon  "))    
```

```python
def maxProfit(nums):
    # https://leetcode.com/problems/best-time-to-buy-and-sell-stock/
    l, r = 0, 1
    mp = 0

    while r < len(nums):
        if nums[r] > nums[l]:
            mp = max(nums[r] - nums[l], mp)
            r = r + 1
        else:
            l = r
            r = r + 1
    return mp    

print(maxProfit([7, 10]))    
```

```python
def kidsWithCandies(candies, extraCandies):
    # https://leetcode.com/problems/kids-with-the-greatest-number-of-candies/description/
    res = []
    maxCandy = max(candies)

    for i in candies:
        if i+extraCandies >= maxCandy:
            res.append(True)
        else:
            res.append(False)
    return res

print(kidsWithCandies([2,3,5,1,3], 3))
```

```python
def findTheDifference(s, t):
    # https://leetcode.com/problems/find-the-difference/
    s_map = {}
    for i in s:
        if i not in s_map:
            s_map[i] = 1
        else:
            s_map[i] += 1
    for j in t:
        if j not in s_map or s_map[j] == 0:
            return j
        s_map[j] -= 1
    return ""               

print(findTheDifference("abcd","abcde"))
```

```python
def findPermutationDifference(s: str, t: str) -> int:
    # https://leetcode.com/problems/permutation-difference-between-two-strings/
    s_map = {v:k for k, v in enumerate(s)}
    diff = 0
    for i in range(len(t)):
        diff += abs(s_map[t[i]] - i)
    return diff   

print(findPermutationDifference('abcde','edbac'))  
```

```python
def isPalindrome(s):
    # https://leetcode.com/problems/valid-palindrome/
    temp = ''.join(i for i in s.lower() if i.isalnum())
    l, r = 0 , len(temp) - 1
    while l < r:
        if temp[l] != temp[r]:
            return False
        l += 1
        r -= 1
    return True    


print(isPalindrome("A man, a plan, a canal: Panama"))
```

```python
def reverseString(s):
    # https://leetcode.com/problems/reverse-string/
    l , r = 0 , len(s) - 1

    while l < r:
        s[l] ,s[r] = s[r], s[l]
        l += 1
        r -= 1
    return s

print(reverseString(['h','e','l','l','o']))    

```

```python
def moveZeroes(nums):
    # https://leetcode.com/problems/move-zeroes/
    pos = 0

    for i in range(len(nums)):
        if nums[i] != 0:
            nums[pos] = nums[i]
            pos += 1

    while pos < len(nums):
        nums[pos] = 0
        pos += 1        
    return nums

def moveZeroes2(nums):
    pos = 0

    for i in range(len(nums)):
        if nums[i] != 0:
            nums[i], nums[pos] = nums[pos], nums[i]
            pos += 1  
    return nums      

print(moveZeroes([0,1,0,3,12]))         
print(moveZeroes2([0,1,0,3,12]))         


```

```python
def bSearch(nums, target):
    # https://leetcode.com/problems/binary-search/
    l, r = 0, len(nums) - 1

    while l <= r:
        mid = (l + r) // 2
        if nums[mid] == target:
            return mid
            break
        elif nums[mid] > target:
            r = mid - 1
        else:
            l = mid + 1
    return -1

print(bSearch([-1,0,3,5,9,12], 9))        
```

```python
def lengthOfLongestSubstring(s):
    # https://leetcode.com/problems/longest-substring-without-repeating-characters/
    resSet = set()
    l = 0
    maxLen = 0

    for r in range(len(s)):
        if s[r] in resSet:
            resSet.remove(s[l])
            l = l + 1
        else:
            resSet.add(s[r])
            maxLen = max(maxLen, len(resSet))
    return maxLen

def lengthOfLongestSubstring2(s: str) -> int:
    charSet = set()
    left = 0
    result = 0
    
    for right in range(len(s)):
        
        while s[right] in charSet:
            charSet.remove(s[left])
            left += 1
        
        charSet.add(s[right])
        
        result = max(result, right - left + 1)
    
    # Return the maximum length of substring found
    return result

print(lengthOfLongestSubstring("pwwkew"))
print(lengthOfLongestSubstring2("pwwkew"))
```

```python
from collections import Counter
def uncommonFromSentences(s1, s2):
    # https://leetcode.com/problems/uncommon-words-from-two-sentences/
    combined = s1.split() + s2.split()
    cntMap = Counter(combined)

    return [res for res in cntMap if cntMap[res]==1]


print(uncommonFromSentences("this apple is sweet", "this apple is sour"))
```

```python
def reformatDate(date):
    mBank = {
        "Jan": "01", "Feb": "02", "Mar": "03", "Apr": "04", "May": "05", "Jun": "06",
        "Jul": "07", "Aug": "08", "Sep": "09", "Oct": "10", "Nov": "11", "Dec": "12"
    }
    day, month, year = date.split()
    day = day[:-2].zfill(2)
    month = mBank[month]

    return f"{year}-{month}-{day}"

print(reformatDate('20th Oct 2052'))
        

```

```python
def validParentheses(s):
    # https://leetcode.com/problems/valid-parentheses/
    pMap = {')':'(','}':'{',']':'['}
    res = []
    for i in s:
        if i in pMap.values():
            res.append(i)
        else:
            if len(res) > 0 and pMap[i] == res[-1]:
                res.pop()
            else:
                return False
    return len(res) == 0

print(validParentheses('()'))        
```

```python
from collections import Counter
def countCharacters(words, chars):
    # https://leetcode.com/problems/find-words-that-can-be-formed-by-characters/description/
    charCount = Counter(chars)
    totLength = 0

    for word in words:
        wordsCount = Counter(word)

        if all(charCount[char] <= wordsCount[char] for char in wordsCount):
            totLength += len(word)
    return totLength

print(countCharacters(["cat","bt","hat","tree"], "atach"))    
```

```python
# https://leetcode.com/problems/subarray-sum-equals-k/description/

def subarraySum(nums, k):
    cnt = 0

    for i in range(len(nums)):
        curSum = 0

        for j in range(i, len(nums)):
            curSum += nums[j]
            if curSum == k:
                cnt += 1
    return cnt 

print(subarraySum([1,1,1], 2))           
            
```

```python
# https://leetcode.com/problems/consecutive-characters/description/

def maxPower(s):
    maxLength = 1
    curLength = 1

    for i in range(1, len(s)):
        if s[i] == s[i-1]:
            curLength += 1
        else:
            curLength = 1
        maxLength = max(maxLength, curLength)
    return maxLength

print(maxPower('leetcode'))        
```

```python
# https://leetcode.com/problems/find-common-characters
from collections import Counter

def commonChars(words):
    # Initialize with the first word's character count
    wrd = Counter(words[0])
    
    # Find the intersection with the rest of the words
    for word in words[1:]:
        wrd = wrd & Counter(word)  # '&=' takes the minimum count for common characters
    
    return list(wrd.elements())

def commonChars2(words):
    common = {}
    res = []
    for i in words[0]:
        if i not in common:
            common[i] = 1
        else:
            common[i] += 1

    for wrd in words[1:]:
        wrdMap = {}
        for w in wrd:
            if w not in wrdMap:
                wrdMap[w] = 1
            else:
                wrdMap[w] += 1

        for k in list(common.keys()): 
            if k in wrdMap:
                common[k] = min(common[k], wrdMap[k])
            else:
                common.pop(k)

    for k, v in common.items():
        res.extend(k*v) 

    return res


words = ["bella","label","roller"]

print(commonChars(words))    
print(commonChars2(words))    
```

```python
# https://leetcode.com/problems/maximum-average-subarray-i/

def findMaxAverage(nums, k):
    curSum = sum(nums[:k])
    maxSum = curSum

    for i in range(k, len(nums)):
        curSum = curSum - nums[i - k] + nums[i]
        maxSum = max(maxSum, curSum)
    return maxSum/k

nums = [1,12,-5,-6, 50, 3]
k = 4
print(findMaxAverage(nums, k))

# 💥 Time Complexity: O(n)
# 💥 Space Complexity: O(1)
```

```python
'''
Q - list of Books prices  = [10,15,2,3,4,25]
Max budget = 25 

Returns maximum no of books that you can buy using this budget
'''

def books(nums, target):
    bookPrice = 0
    bookCount = 0
    for i in sorted(nums):
        if i + bookPrice <= target:
            bookPrice += i
            bookCount += 1
    return bookCount

print(books([10,15,2,3,4,25], 25))    

```
